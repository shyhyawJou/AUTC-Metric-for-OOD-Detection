# __Torch Version__

```python
from time import time_ns
import numpy as np
from matplotlib import pyplot as plt
import torch
from metric.torch import autc
```

# Test Case #1

```python
DEVICE = 'cpu'
id_scores = torch.tensor(np.load('data/id_1.npy'), device=DEVICE)
ood_scores = torch.tensor(np.load('data/ood_1.npy'), device=DEVICE)
y = torch.tensor([0] * len(id_scores) + [1] * len(ood_scores), device=DEVICE)
score = torch.cat((id_scores, ood_scores))
```

```python
autc_score, fpr, fnr, thresh = autc(y, score)
autc_score = autc_score.cpu().numpy()
fpr = fpr.cpu().numpy()
fnr = fnr.cpu().numpy()
thresh = thresh.cpu().numpy()
```

```python
plt.plot(thresh, fnr)
plt.plot(thresh, fpr)
plt.fill_between(thresh, 0, fnr, alpha=0.5, label="FNR")
plt.fill_between(thresh, 0, fpr, alpha=0.5, label="FPR")
plt.legend(loc="upper center")
plt.title(f'AUTC: {autc_score:.4f}', weight='bold')
plt.show()
```
